# Criando gráficos a partir de um dataframe

O objetivo aqui é plotar gráficos de vendas de quatro lojas fictícias (A, B, C e D).
Os dados são referentes às vendas por mês no ano de 2022.

```python
import pandas as pd
import matplotlib.pyplot as plt

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# criação de um dataframe com as informações de vendas correspondentes a cada loja
df = pd.DataFrame(vendas_2022, index=lojas)
```

```python
# criando uma figura
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D')
axs[0, 0].plot(df.loc['A'])
axs[0, 0].set_title('Vendas na loja A')
axs[0, 1].plot(df.loc['B'])
axs[0, 1].set_title('Vendas na loja B')
axs[1, 0].plot(df.loc['C'])
axs[1, 0].set_title('Vendas na loja C')
axs[1, 1].plot(df.loc['D'])
axs[1, 1].set_title('Vendas na loja D')

for ax in axs.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Número de vendas')

plt.show()
```

# Customizando gráficos com Matplotlib

A biblioteca Matplotlib oferece diversas opções de customização da figura acima.
Quando é desejável que as modificações sejam aplicadas a todos os gráficos da figura, é possível resumir o código por meio de um loop iterando sobre todos os subplots da figura.

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# ajustando o espaço entre os subplots
plt.subplots_adjust(wspace=0.4, hspace=0.5)

# título geral
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=16)

# lista de cores a ser utilizada
cores = ['navy', 'slateblue', 'rebeccapurple', 'purple']

# plotar os gráficos por meio de um loop
for i, ax in enumerate(axs.flat):
    ax.plot(df.loc[df.index[i]], color=cores[i], lw=3)
    ax.set_title(f'Vendas na loja {df.index[i]}', loc='center', fontsize=14)
    ax.set_xlabel('Mês', fontsize=14)
    ax.set_ylabel('Número de vendas', fontsize=14)
    ax.set_ylim(0,450)
    ax.yaxis.set_major_locator(plt.MultipleLocator(50))
    ax.tick_params(labelsize=11)
    ax.grid(linestyle='--', color='lightgrey')

plt.show()
```
